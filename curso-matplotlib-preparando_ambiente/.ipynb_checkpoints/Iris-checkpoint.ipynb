```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('C:/Alura/curso-matplotlib-preparando_ambiente/iris.csv')
```

```python
df.sample(50)
```

```python
fig = plt.figure(figsize=(12,6))

ax=fig.add_axes([0, 0, 1, 1])
ax.scatter(df['comprimento_sépala'], df['largura_sépala'], color='k')

ax.set_title('Gráfico de Dispersão', fontsize=15)
ax.set_xlabel('Comprimento da sépala', fontsize=12)
ax.set_ylabel('Largura da sépala', fontsize = 12)
```

```python
fig = plt.figure(figsize=(12,6))

ax=fig.add_axes([0, 0, 1, 1])

cores={'Iris-virginica':'m', 'Iris-setosa':'b', 'Iris-versicolor':'y'}
marcadores={'Iris-virginica':'*', 'Iris-setosa':'+', 'Iris-versicolor':'^'}#adicionar marcadores

for especie in df['espécie'].unique():
    tmp=df[df['espécie'] == especie]
    ax.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
              color=cores[especie], marker=marcadores[especie],
              s=150)

ax.set_title('Gráfico de Dispersão', fontsize=15)
ax.set_xlabel('Comprimento da sépala', fontsize=12)
ax.set_ylabel('Largura da sépala', fontsize = 12)
ax.legend(cores, fontsize = 12)
ax.tick_params(labelsize=15)#aumenta o tamanho dos ticks
```

```python
fig = plt.figure(figsize=(5,4))
ax=fig.add_axes([0, 0, 1, 1])



ax.hist(df['comprimento_pétala'], bins=20)#bins número de colunas
ax.set_title('Histograma Pétalas', fontsize=15, pad=10)
ax.set_xlabel('Comprimento da pétala', fontsize=15)
ax.grid(True)
```

```python
fig = plt.figure(figsize=(5,4))
ax=fig.add_axes([0, 0, 1, 1])

mu, sigma = (df['comprimento_pétala'].mean(), df['comprimento_pétala'].std())


ax.hist(df['comprimento_pétala'], bins=20)#bins número de colunas
ax.set_title('Histograma Pétalas', fontsize=15, pad=10)
ax.set_xlabel('Comprimento da pétala', fontsize=15)
ax.grid(True)
ax.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),#com a \ muda para simbolo
           xy = (4.5, 20), fontsize=15)
#média
ax.axvline(mu, color = 'k', linestyle='--')
ax.annotate('média', xy=(mu-1.3, 28), fontsize=20)
#mediana
ax.axvline(df['comprimento_pétala'].median(), color = 'r', linestyle='--')
ax.annotate('mediana', xy=(df['comprimento_pétala'].median(), 31), fontsize=20)

fig.savefig('histograma.jpeg', bbox_inches='tight')
```
